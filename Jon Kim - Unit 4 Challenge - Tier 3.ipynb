**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

```python
# Create function that calculate the price ratio for each borough

def create_price_ratio(df, year1, year2):
    """
    For a specific borough in the dataframe, take the average prices in year1 and year2, and
    calculate & return the ratio between them.
    
    df must be a twice-indexed data frame with borough as first index and date as the second index.
    """
    
    # Unique Boroughs
    global unique_list
    unique_list =[]
    
    for b in tuple(df.iterrows()):
        if b[0][0] not in unique_list:
            unique_list.append(b[0][0])
    
    # Create list with the ratios for each borough
    global borough_ratio_list
    borough_ratio_list = []

    c_borough = ''
    c_year1_price = 0
    c_year2_price = 0
    c_price_ratio = 0
    for b in unique_list:
        for row in tuple(df.iterrows()):
            c_borough = b
            if row[0][1]== year1 or row[0][1]== year2:
                if row[0][0] == b and row[0][1]== year1:
                    c_year1_price = row[1][0]
                elif row[0][0] == b and row[0][1]== year2:
                    c_year2_price = row[1][0]
                    if c_year1_price != 0 and c_year2_price !=0:
                        c_price_ratio = c_year2_price / c_year1_price
                        borough_ratio_list.append(tuple((c_borough, year1, c_year1_price, year2, c_year2_price, \
                                                         c_price_ratio)))
                    else:
                        borough_ratio_list.append(tuple((c_borough, year1, year2, None)))

    # Convert new borough ratio list into a dataframe
    global b_ratio_df
    b_ratio_df = pd.DataFrame(borough_ratio_list, columns=['Borough','Year 1', 'Year 1 Avg Price', 'Year 2', \
                                                           'Year 2 Avg Price','Price Ratio'])
    
    #display the new dataframe with price ratios
    return(b_ratio_df)
```

```python
# Use function to output list of price ratios by borough

properties19_price_ratios = create_price_ratio(properties19,1998,2018)
display(properties19_price_ratios)
```

```python
# Sort new price ratio dataframe to see the boroughs with the highest ratios

properties19_price_ratios_sort = properties19_price_ratios.sort_values('Price Ratio', ascending=False)
display(properties19_price_ratios_sort)
```
